# Bayesian Core Analysis: Hypothesis Tests and Prior Checks

This notebook re-runs the main hypothesis tests using the Bayesian-derived outcome and interprets the null hypotheses under the stricter synthetic-data policy.


```python
from pathlib import Path
from itertools import combinations

import numpy as np
import pandas as pd
from scipy import stats
from statsmodels.stats.oneway import anova_oneway
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from statsmodels.stats.multitest import multipletests
from IPython.display import display, Markdown

ALPHA = 0.05
```

```python
project_root = Path.cwd()
while not (project_root / 'data').exists() and project_root != project_root.parent:
    project_root = project_root.parent

data_path = project_root / 'data' / 'processed' / 'OECD_FDI_Except_Pointless.csv'
df = pd.read_csv(data_path, sep=';', encoding='latin1', engine='python', on_bad_lines='skip')
df['OBS_VALUE_NUM'] = pd.to_numeric(df['OBS_VALUE'].astype(str).str.replace('.', '', regex=False), errors='coerce')
df['obs_value_imputed'] = df['OBS_VALUE_NUM'].fillna(df['OBS_VALUE_NUM'].median())
df['obs_value_signed_log'] = np.sign(df['obs_value_imputed']) * np.log1p(np.abs(df['obs_value_imputed']))

global_mean = df['obs_value_signed_log'].mean()
global_var = df['obs_value_signed_log'].var(ddof=1)
def empirical_bayes_mean(data, group_col, value_col='obs_value_signed_log'):
    grouped = data.groupby(group_col)[value_col]
    stats_df = grouped.agg(['mean', 'count', 'var']).reset_index()
    between_var = max(stats_df['mean'].var(ddof=1), 1e-6)
    within_var = max(stats_df['var'].fillna(global_var).mean(), 1e-6)
    stats_df['shrinkage_weight'] = (stats_df['count'] * between_var) / (stats_df['count'] * between_var + within_var)
    stats_df['posterior_mean'] = stats_df['shrinkage_weight'] * stats_df['mean'] + (1 - stats_df['shrinkage_weight']) * global_mean
    return stats_df[[group_col, 'posterior_mean']]
for group_col, new_col in [('REF_AREA', 'bayes_ref_area'), ('ACTIVITY', 'bayes_activity'), ('TYPE_ENTITY', 'bayes_type_entity')]:
    post = empirical_bayes_mean(df, group_col).rename(columns={'posterior_mean': new_col})
    df = df.merge(post, on=group_col, how='left')
df['obs_value_bayes_selected_signed_log'] = df[['bayes_ref_area', 'bayes_activity', 'bayes_type_entity']].mean(axis=1)
df['bayes_best_probability'] = 1 - ((df['bayes_ref_area'] - df['obs_value_signed_log']).abs() / (1 + df['obs_value_signed_log'].abs()))
df['bayes_best_probability'] = df['bayes_best_probability'].clip(lower=0, upper=1)
df[['obs_value_bayes_selected_signed_log', 'bayes_best_probability']].head()
```

## Correlation and Association Tests


```python
pearson_r, pearson_p = stats.pearsonr(df['TIME_PERIOD'], df['obs_value_bayes_selected_signed_log'])
spearman_r, spearman_p = stats.spearmanr(df['TIME_PERIOD'], df['obs_value_bayes_selected_signed_log'])
corr_df = pd.DataFrame([
    {'method': 'Pearson', 'coefficient': pearson_r, 'p_value': pearson_p},
    {'method': 'Spearman', 'coefficient': spearman_r, 'p_value': spearman_p},
    {'method': 'Absolute difference', 'coefficient': abs(pearson_r - spearman_r), 'p_value': np.nan},
])
corr_df.round(6)
```

```python
def cramers_v(table):
    chi2 = stats.chi2_contingency(table, correction=False)[0]
    n = table.to_numpy().sum()
    r, k = table.shape
    return np.sqrt(chi2 / (n * (min(r, k) - 1)))

chi_table = pd.crosstab(df['TYPE_ENTITY'], df['MEASURE_PRINCIPLE'])
chi2, chi_p, dof, expected = stats.chi2_contingency(chi_table)
chi_df = pd.DataFrame([{'chi2': chi2, 'p_value': chi_p, 'dof': dof, 'cramers_v': cramers_v(chi_table)}])
chi_df.round(6)
```

## Prior Assumption Checks Before Group Mean Comparisons


```python
variance_rows = []
for group_col in ['TYPE_ENTITY', 'MEASURE_PRINCIPLE']:
    groups = [g['obs_value_bayes_selected_signed_log'].to_numpy() for _, g in df.groupby(group_col)]
    lev = stats.levene(*groups, center='median')
    bart = stats.bartlett(*groups)
    variance_rows.append({'group_variable': group_col, 'levene_p': lev.pvalue, 'bartlett_p': bart.pvalue, 'equal_variance_supported': (lev.pvalue >= ALPHA) and (bart.pvalue >= ALPHA)})
variance_df = pd.DataFrame(variance_rows)
variance_df.round(6)
```

```python
def pairwise_welch(data, outcome, group):
    results = []
    groups = {k: v[outcome].dropna().to_numpy() for k, v in data.groupby(group)}
    pairs = list(combinations(groups.keys(), 2))
    raw_p, raw_stats = [], []
    for g1, g2 in pairs:
        stat, p_value = stats.ttest_ind(groups[g1], groups[g2], equal_var=False)
        raw_p.append(p_value)
        raw_stats.append((g1, g2, stat, p_value))
    reject, p_adj, _, _ = multipletests(raw_p, alpha=ALPHA, method='holm')
    for (g1, g2, stat, p_value), rj, padj in zip(raw_stats, reject, p_adj):
        results.append({'group_1': g1, 'group_2': g2, 'welch_t': stat, 'raw_p': p_value, 'holm_p': padj, 'reject_h0': bool(rj)})
    return pd.DataFrame(results)

anova_rows = []
posthoc_tables = {}
for group_col in ['TYPE_ENTITY', 'MEASURE_PRINCIPLE']:
    equal_variance = bool(variance_df.loc[variance_df['group_variable'] == group_col, 'equal_variance_supported'].iloc[0])
    if equal_variance:
        res = anova_oneway(df['obs_value_bayes_selected_signed_log'], groups=df[group_col], use_var='equal')
        posthoc_tables[group_col] = pairwise_tukeyhsd(df['obs_value_bayes_selected_signed_log'], df[group_col]).summary()
        method = 'ANOVA + Tukey'
    else:
        res = anova_oneway(df['obs_value_bayes_selected_signed_log'], groups=df[group_col], use_var='unequal')
        posthoc_tables[group_col] = pairwise_welch(df, 'obs_value_bayes_selected_signed_log', group_col)
        method = 'Welch ANOVA + pairwise Welch-Holm'
    anova_rows.append({'group_variable': group_col, 'selected_method': method, 'F_statistic': res.statistic, 'p_value': res.pvalue})
anova_df = pd.DataFrame(anova_rows)
anova_df.round(6)
```

```python
display(Markdown('### TYPE_ENTITY post-hoc'))
display(posthoc_tables['TYPE_ENTITY'])
display(Markdown('### MEASURE_PRINCIPLE post-hoc'))
display(posthoc_tables['MEASURE_PRINCIPLE'])
```
